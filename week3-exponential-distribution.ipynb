```python
import sympy as sym
from sympy import Symbol,exp,pprint
import sympy.plotting as syp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a=Symbol('lambda')
x=Symbol('x')
e=Symbol('e')
```

```python
myf_1=a*sym.exp(-x*a) #Exponential Distribution formula
syp.plot(myf_1.subs({e:1,a:2}),(x,0,10),title="Exponential Distribution")
plt.show()
```

```python
x_values=[]
y_values=[]
for value in range(-10,10):
    y=myf_1.subs({a:5,x:value}).evalf()
    y_values.append(y)#Listeye alma iÅlemi
    x_values.append(value)
plt.plot(x_values,y_values)
plt.show()
```
